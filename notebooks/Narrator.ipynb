```python
class NarratorAgent:
    def run(self, raw_caption):
        sections = [
            "IMAGE TYPE",
            "UTERINE MORPHOLOGY",
            "MYOMETRIAL ASSESSMENT",
            "JUNCTIONAL ZONE",
            "ENDOMETRIUM",
            "VISIBLE ANNOTATIONS",
            "IMPRESSION"
        ]

        output = {}

        for section in sections:
            if section in raw_caption:
                start = raw_caption.find(section)
                output[section] = raw_caption[start:start+200]
            else:
                output[section] = "Not clearly visualised"

        output["CONFIDENCE"] = 0.75
        return output
```
